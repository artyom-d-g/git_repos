# 27. Какие актеры чаще всего снимаются в одном фильме вместе?


```python
#функция для сортировки актеров в единообразном порядке. Откуда-то спозаимствовано
#Нужна, чтобы не было лишних пар, а только единообразные
#Возможно, это делается через что-то иное вместо combinations, но
# раз combinations, то она мне нужна. А еще думать - пока не успеваю.
def sortByAlphabet(inputStr):    
        return inputStr[0]

    
#функция, создающая пары актеров из списка
def pairs_maker(x):
    x.sort(key=sortByAlphabet)
    x = itertools.combinations(x,2)
    return list(x)


actors_x_n_filmns = data['cast'].explode()    
actors_count = actors_x_n_filmns.value_counts()    #получаем количество фильмов, в которых снялся каждый
#искать топ пары будем среди 20% самых часто снимающихся актеров
#переводим сразу в список из серии
top_80 = actors_count[actors_count >= np.percentile(actors_count, 80)].index.tolist()    

work_data = data    # чтобы не портить исходник

#Переводим значения из списков в единичные; со списками .isin не работает
#Кстати, буду признаетелен, если расскажете, как осуществить сортировку
#по условию вхождения списков в Серии Дата фрейма в другой список.
k_data = work_data.explode('cast')

data_80_exld = k_data[k_data['cast'].isin(top_80)]    #Обрезаем наш Дф, чтобы программе легче было работать

# А теперь возвращаем значения в колонках к спискам.
# Кстати, был бы признателен за подсказку функции, которая так делает
# то есть осущствляет обратное действие функции explode
data_80 = data[data['imdb_id'].isin(data_80_exld['imdb_id'])]
data_80_pairs = data_80['cast'].apply(pairs_maker)    #Превращаем списки актеров по каждому фильму во все возможные пары
data_80_pairs_expl = data_80_pairs.explode()    #Получаем Серию, где пары актеров разделены по одной паре на строку * на N фильмов

#Считаем наконец
answer_dict = {}
for i in data_80_pairs_expl:
    if i in answer_dict:
        answer_dict[i] += 1
    else:
        answer_dict[i] = 1
        
Counter(answer_dict).most_common(4)    #Берем 4 топ пары
```
